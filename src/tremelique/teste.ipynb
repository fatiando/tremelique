```python
import numpy as np
import matplotlib.pyplot as plt
from _utils import create_homogeneous_model 
from _acoustic import Acoustic
from _wavelets import RickerWavelet

shape = (300, 400)
spacing = 5
propriedades_acusticas = {"velocity": 1500, "density": 1000}

model = create_homogeneous_model(shape, propriedades_acusticas)


sim = Acoustic(
    velocity=model["velocity"], 
    density=model["density"], 
    spacing=spacing,
    verbose=True
)

sim.add_point_source((0, shape[1]//2), RickerWavelet(1,60))

sim.run(400)

anim = sim.animate(every=10, embed=True, cutoff=0.5, dpi=50)
anim 
```
