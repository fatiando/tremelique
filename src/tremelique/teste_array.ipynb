```python
import numpy as np
import matplotlib.pyplot as plt
import bordado as bd

from _utils import create_homogeneus_model_xarray
from _acoustic import Acoustic

#region = [xmin, xmax, zmin, zmax] (metros?)
region = [0, 3000, 0, 2000]
shape = (201, 301)
props = {
    "velocity": 1500.00,
    "density": 1000.00}
print(f'Região:{region}')
print(f'Shape: {shape}')
print(f'Propriedades: {props}')

modelo = create_homogeneus_model_xarray(region,shape,props)
print(modelo)
print(f'dz = {modelo.attrs["dz"]} metros \ndx = {modelo.attrs["dx"]} metros')

fig, (ax1, ax2) = plt.subplots(1 , 2, figsize=(14,5))
modelo.velocity.plot(
    ax=ax1,
    yincrease=False #faz o eixo Z(profundidade) aumentar para baixo
)
ax1.set_title("Grid 'velocity'")

modelo.density.plot(
    ax=ax2,
    yincrease=False)
ax2.set_title("Grid 'density'")
plt.tight_layout(rect=[0,0.3,1,0.95])
plt.show()
```
