```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# Tremelique
from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
from tremelique._utils import model
```

```python
region = [0, 2000, 0, 1500]
spacing = 5
```

```python
layers = [
    {"z": 0, "velocity": 1500.0, "density": 1000.0},
    {"z": 500, "velocity": 2000.0, "density": 1500.0}
]
model = model(region,spacing,layers)
```

```python
sim = Acoustic(model)
```

```python
sim.add_point_source_physical((1000, 0),RickerWavelet(1, 60)) 
#diferente do trabalho com index, o add_point_source_physical trabalha com (nx, nz), no caso (posição em x, posição em z), mas converte para indice e passa para o tremelique para não quebrar o codigo
```

```python
sim.run(700)
```

```python
sim.animate_physical(every=10, embed=True, dpi=50, cutoff=0.5)
```
