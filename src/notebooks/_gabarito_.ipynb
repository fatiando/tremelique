```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from tremelique._utils import model
from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
```

```python
region = [0, 3000, 0, 2000] #xmin, xmax, zmin, zmax
spacing = 5
props = {"velocity":1500,"density":1000}

model = model(region, spacing, properties=props)
```

```python
sim = Acoustic(model)
sim.add_point_source_physical((1500,0), RickerWavelet(1,60))
```

```python
sim.run(400)
```

```python
anim = sim.animate_physical(every=10, embed=True, cutoff=0.5)
anim
```
