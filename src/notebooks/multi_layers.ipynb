```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
from tremelique._utils import layered_model
```

```python
region = [0,2000, 0, 1500]
spacing = 5
```

```python
layers = [
    {"z": 0,"properties": { "velocity": 1500.0, "density": 1000.0}},
    {"z": 500,  # equivalente ao índice 100
        "properties": {
            "velocity": 2000.0,
            "density": 1500.0,
        },
    },
]
```

```python
model = layered_model(region=region, spacing=spacing, layers=layers)
model
```

```python
sim = Acoustic(model=model, verbose=True)
```

```python
sim.add_point_source((0, 200), RickerWavelet(1,60))
```

```python
sim.run(700)
```

```python
sim.animate(every=10, embed=True, dpi=50, cutoff=0.5)
```
