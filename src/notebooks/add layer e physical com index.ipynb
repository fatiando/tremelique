```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
from tremelique._utils import layered_model, homogeneous_model_xarray, add_layer_xarray
```

```python
model = homogeneous_model_xarray(
    region=(0, 2000, 0, 1500),
    spacing=5,
    properties={"velocity": 1500, "density": 1000},
)
```

```python
sim = Acoustic(model=model)
```

```python
sim.add_point_source((0, 200), RickerWavelet(1,60)) #o add_point_source usa ((z, x), wavelet); já o add_point_source_physical ((x, z), wavelet)
```

```python
sim.run(700)
```

```python
sim.animate(every=10, embed=True, dpi=50, cutoff=0.5)
```

```python
model = add_layer_xarray(
    model,
    z_top=400,
    z_bottom=900,
    properties={"velocity": 3000, "density": 2400},
)
```

```python
sim2 = Acoustic(model)
sim2.add_point_source_physical((1000, 0), RickerWavelet(1,60)) #o add_point_source usa ((z, x), wavelet); já o add_point_source_physical ((x, z), wavelet)
```

```python
sim2.run(700)
```

```python
sim2.animate_physical(every=10, embed=True, dpi=50, cutoff=0.5)
```
