```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
from tremelique._utils import model
```

```python
region = [0, 3000, 0, 2000]
spacing = 5

model = model(region, spacing, velocity=1500, density=1000)
```

```python
sim = Acoustic(model, dt=0.001)
sim.add_point_source((0,300), RickerWavelet(1,60))
sim.run(400)

sim.animate(every=10, embed=True, dpi=50, cutoff=0.5)
```
