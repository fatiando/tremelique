```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

from tremelique._acoustic import Acoustic
from tremelique._wavelets import RickerWavelet
from tremelique._utils import model
```

```python
region = [0, 27000, 0, 7500]
spacing = 30.0
shape = (250, 900)   # (nz, nx) -> shape do notebook

nz, nx = shape
vp1 = 3000.0
vp2 = 4000.0
rho1 = 2000.0
rho2 = 2300.0

interface_index = 80
interface_depth = interface_index * spacing
```

```python
layers = [{"z": 0.0,"velocity": 3000,"density": 2000},{"z": 2400,"velocity": 4000,"density": 2300}]
model = model(region=region, spacing=spacing, layers=layers,)
```

```python
model
```

```python
sim = Acoustic(model=model)
```

```python
fonte = (0, nx // 5) #(nz, nx)
fonte_physical = (5400, 0) #(nx, nz)
print("Coordenadas da fonte:")
print(f"  x = {fonte_physical[0]} m")
print(f"  z = {fonte_physical[1]} m")
sim.add_point_source_physical((5400,0), RickerWavelet(1, 30))
```

```python
sim.run(2000)
```
